GridSearch
====

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
```

```python
dataset = pd.read_csv('./unified.csv')
dataset['Data'] = pd.to_datetime(dataset['Data'], format="%Y/%m/%d")
dataset.head(10)
```

```python
dataset.info()
```

```python
classes = dataset[['Precipitacao']].values
classes = np.squeeze(np.asarray(classes))

dataset.drop(['Precipitacao'], axis=1, inplace=True)

data = dataset.iloc[:, range(len(dataset.columns))].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, classes, test_size=.8)
```

```python
svc_parameters = [
    { 'kernel': ['rgf'], 'C': [1, 10, 100, 1000] },
    { 'kernel': ['linear'], 'C': [1, 10, 100, 1000] },
    { 'kernel': ['poly'], 'C': [1, 10, 100, 1000] },
    { 'kernel': ['sigmoid'], 'C': [1, 10, 100, 1000] }
]

scores = ['precision', 'recall']
```

```python
for score in scores:
    print ('Hyper-Parameters for %s' % score)
    
    grid_search = GridSearchCV(
        estimator=SVC(), 
        param_grid=svc_parameters, 
        scoring='%s_macro' % score, 
        n_jobs=-1)
    
    grid_search.fit(x_train, y_train)
    print('Best parameters set found on training data:')
    print()
    print(grid_search.best_params_)
    print()
    
    y_true, y_pred = y_test, grid_search.predict(x_test)
    print('Fit results:')
    print()
    print(classification_report(y_true, y_pred))
    print()
```
